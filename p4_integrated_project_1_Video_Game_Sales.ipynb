UPD. I was wrong, that didn't make it easier. Now I remember dt.year doesn't prevent that because it becomes int. So I patched the function to round the value down.

```python
data['year_of_release'] = pd.to_datetime(data['year_of_release'], format = '%Y')
data['year_of_release'] = data['year_of_release'].dt.year
```

So these are the games with missing values in year:

```python
games_wo_year = data.loc[data['year_of_release'].isnull()]['name'].unique()
```

We can write a function that looks up the game's median value for year if it appears on other platforms, or, in case it doesn't, fill it in with the median year for the platform:

```python
for game in games_wo_year:
    game_year_median = data.loc[data['name'] == game]['year_of_release'].median()
    if not pd.isna(game_year_median): # this is to check in case there is just one release: in that case we'll fill in by platform
        data.loc[data['name'] == game, 'year_of_release'] = data['year_of_release'].fillna(round(game_year_median)) # rounding to the earliest year of release for that game in case it's not a round year
    else:
        data.loc[data['name'] == game, 'year_of_release'] = data['year_of_release'].fillna(data.groupby('platform')['year_of_release'].transform('median'))

        
```

### User and professional reviews' effect on sales

Let's take our five most-performing platforms in terms of sales and see if those are in any way effected by the reviews that critics and users leave. We can build a function to test it on the top-performing platform and then apply it on its competitor platforms.

We have to remember that we have missing values for both critic and user reviews. We have to drop them for this to work, but we should keep in mind that in this case we are measuring the effect (or lack of it) of reviews only where they are present, and that it is NOT the same thing as measuring the mere presence or absence of reviews on sales.

```python
fresh_data = new_data.dropna(subset = ['critic_score', 'user_score'])
```

```python
fresh_data.info()
```

```python
def review_effect_plot(platform, review): # 'user_score' or 'critic_score'
    subset = fresh_data.loc[fresh_data['platform'] == platform][['total_sales', review]]
    #if review == 'user_score':
    #    subset = subset.loc[subset['user_score'] != "tbd"]
    #    subset['user_score'] = subset['user_score'].astype(float)
        
    subset.plot(kind = 'scatter', x = review, y = 'total_sales', figsize = (10, 6))
    plt.title(f'Sales on {platform} by {review}')
    plt.show()
```

```python
def show_corr(platform, review):
    subset = fresh_data.loc[fresh_data['platform'] == platform][['total_sales', review]]
    #if review == 'user_score':
    #    subset = subset.loc[subset['user_score'] != "tbd"]
    #    subset['user_score'] = subset['user_score'].astype(float)
        
    p_corr = subset['total_sales'].corr(subset[review])
    print(p_corr)
    if p_corr > 0.7:
        print(f'There is a positive correlation between sales and {review} for games on {platform}')
    else:
        print(f'There is no correlation between sales and {review} for games on {platform}')
```

```python
review_effect_plot('PS4', 'user_score')
```

```python
show_corr('PS4', 'user_score')
```

The negative Pearson coefficient tells that there is negative correlation between the two variables, however it is too close to 0, so one can say it is so weak as if there was no relationship between the variables at all (almost). So, if imagined it was closer to -1 then there would be a strong negative correlation and games with worse reviews would be likely to sell more! A wonder indeed. But that is not the case and we can only see the direction of correlation is negative (such as it is ‚Äì which is weak). How about critics' reviews?

## Create a user profile for each region

Let's do a breakdown for the three main regions in terms of sales: do they differ by platforms and genres? We'll do this using a function. I will round the values for market share for simplicity.

```python
def show_top_platforms(region, parameter): # 'platform' or 'genre'
    top = new_data[[parameter, region]].groupby(parameter).sum().sort_values(by = region, ascending = False).reset_index()
    top['share'] = round((top[region]/top[region].sum()) * 100)
    print(top.head())
    
```

```python
regions = ['na_sales', 'eu_sales', 'jp_sales']
for region in regions:
    print(f'Top five platforms by sales and market share for {region}:')
    print()
    show_top_platforms(region, 'platform')
    print()
```

We can see how Xbox is more popular in the States, with 24 percent of the sales, and the new generation XOne stepping on the heels of both models of PS, where for EU it's the opposite. In both of these markets the sales are dominated by these two gaming stations, and the handheld Nintendo 3 takes smallest share (still, meaning in the top 5) of less than 10%. Alright. The sample sizes are good enough for our test! Let's go:

Because we are comparing independent instances of user score for different platforms between each other and asking whether their mean value is the same, it's appropriate that we use a two-tailed t-test. We are testing if means of two statistical populations are equal. In this case our null hypothesis (H0) is: "The average user rating for (games on) XOne and PC is equal". And so the alternative hypothesis (H1) is: "The average user_score for XOne and PC is not equal".

It is not guaranteed likely that the variance of the two groups (average user scores for the two platforms) is the same. To fool-proof against that possibility, we will include another method inside the function that tests for that. If the variance is different, the function inside the function will pass the respective third parameter to our ttest (by default it assumes that the variance is equal), so that we can switch to perform Welch's ttest, which is more appropriate for when variance is not equal. And we will use the standard alpha of 5% to assess the result.

```python
def t_test(sample_1, sample_2, s1_name, s2_name, alpha = 0.05): # we are just passing the names for extra printing feature at the end
    p_value_levene = st.levene(sample_1, sample_2).pvalue
    if p_value_levene < alpha:
        var_value = False
    else:
        var_value = True
    
    p_value = st.ttest_ind(sample_1, sample_2, equal_var = var_value, nan_policy = 'omit').pvalue
    print(p_value)
    if p_value < alpha:
        print(f'Reject Null Hypothesis: the average user score for {s1_name} and {s2_name} games is different. There is a very tiny probability of randomly getting any other difference between them, therefore we must reject the probability that they can be equal.')
    else:
        print('Cannot reject the Null Hypothesis: there is a relatively high chance of getting similar average user score.')
        
```

```python
t_test(xbox, pc, "Xbox One", "PC")
```

A playah's a playah! On both Xbox and PC. Interesting, so we know that there is high chance that average user review for Xbox and PC games would be the same.
